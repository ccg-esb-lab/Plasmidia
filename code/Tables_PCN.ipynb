## Set up system

```python
from google.colab import auth
auth.authenticate_user()

import gspread
import pandas as pd

from oauth2client.client import GoogleCredentials
gc = gspread.authorize(GoogleCredentials.get_application_default())
```

```python

sp_path="plasmids_in_nature"

worksheet = gc.open(sp_path)
print(worksheet.worksheets())
```

## Table PCN vs Cost

### read table

```python
pcnVsCost_table=worksheet.worksheet("PCNvsCost")
rows=pcnVsCost_table.get_all_values()


pcnCost_df_raw=pd.DataFrame.from_records(rows[:13],columns=rows[0]).drop(0)


pcnCost_df_raw
```

### edit table to latex format

```python
pcnCost_df=pcnCost_df_raw.copy()
textrefs=["\cite{%s}"%x for x in pcnCost_df_raw.Reference]
pcnCost_df.Reference=textrefs

species_it=[]
for x in pcnCost_df_raw.Species:
    xs=x.split(" ")
    print(xs)
    if(len(xs)==2):
        this_name="\\textit{%s %s}"%(xs[0],xs[1])
    elif(len(xs)>2):
        this_name="\\textit{%s %s} %s"%(xs[0],xs[1],xs[2])
    species_it.append(this_name)
pcnCost_df.Species=species_it

#pcnCost_df=pcnCost_df.iloc[:,0:9,11]
#pcnCost_df=pcnCost_df[['Name','Plasmid Type','Species','PCN','PCN SD','Cost','Cost SD','Rel F','Rel F SD','seg rate','doi','Reference']]
pcnCost_df=pcnCost_df[['Name','Plasmid Type','Species','PCN','PCN SD','Cost','Cost SD','Reference']]
pcnCost_df
```

### save csv file


```python
project_path="/content/drive/My Drive/SYNC_Projects/Plasmidia/ms/"   ####  <-------------------- check
table_path=project_path+"data/plasmidsInNature_tables/"
pcnCost_df.to_csv(table_path+"pcnCost_table_latex.csv",sep=",",index=False,index_label=False)

pcnCost_df_raw.to_csv(table_path+"pcnCost_table.csv",sep="\t",index=False,index_label=False)
```

## Table PCN vs Size

### read table

```python

sp_path="plasmids_in_nature"

worksheet = gc.open(sp_path)
print(worksheet.worksheets())
```

```python
pcnVsSize_table=worksheet.worksheet("PCNvsSize")
rows=pcnVsSize_table.get_all_values()
rows=rows[0:17]

pcnSize_df_raw=pd.DataFrame.from_records(rows,columns=rows[0]).drop(0)


pcnSize_df_raw
```

### edit table to latex format

```python
pcnSize_df=pcnSize_df_raw.copy()
textrefs=["\cite{%s}"%x for x in pcnSize_df.Reference]
pcnSize_df.Reference=textrefs

species_it=[]
for x in pcnSize_df.Species:
    xs=x.split(" ")
    print(xs)
    if(len(xs)==2):
        this_name="\\textit{%s %s}"%(xs[0],xs[1])
    elif(len(xs)>2):
        this_name="\\textit{%s %s} %s"%(xs[0],xs[1],xs[2])
    species_it.append(this_name)
pcnSize_df.Species=species_it

texttype=[]
for x in pcnSize_df['Plasmid Type']:
    if(x=='𝛉'):
        x="$\\theta$"
    texttype.append(x)
pcnSize_df['Plasmid Type']=texttype

#𝛉

#pcnCost_df=pcnCost_df.iloc[:,0:9,11]
#pcnCost_df=pcnCost_df[['Name','Plasmid Type','Species','PCN Mean','PCN SD','Size (Kb)','Special Notes','ref','Reference']]
pcnSize_df=pcnSize_df[['Name','Plasmid Type','Species','PCN Mean','PCN SD','Size (Kb)','Reference']]
pcnSize_df
```

### save csv file


```python
project_path="/content/drive/My Drive/SYNC_Projects/Plasmidia/ms/"   ####  <-------------------- check
table_path=project_path+"data/plasmidsInNature_tables/"


pcnSize_df_raw.to_csv(table_path+"pcnSize_table.csv",sep="\t",index=False,index_label=False)
pcnSize_df.to_csv(table_path+"pcnSize_table_latex.csv",sep=",",index=False,index_label=False)
```
